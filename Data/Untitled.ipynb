```python
import pandas as pd
import numpy as np
```

```python
data = pd.read_csv("IMDb_clean.csv")
data = data[['title', 'genre', 'avg_vote']]
```

```python
data.genre = data.genre.apply(lambda x : x.split(', ')[0] if ', 'in x else x)
```

```python
data[data['title'] == 'The Avengers']
```

```python
result = {}
for genre, df in data.groupby('genre'):
    result[genre] = df.sort_values(by='avg_vote', ascending=False).head(10).values.tolist()
```

```python
result
```

```python
import json
with open('top_movies.json', 'w') as f:
    json.dump(result, f)
```
